```python
df.isnull().sum()
```

```python
df.isnull()
```

```python
#useful to find the null values in the data
df.isnull().sum()
```

```python
#this query is useful in entire EDA for finding null values of a particular column by
# traversing a temp variable features in the columns of data to find the null values
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
#By using heatmap we can plot the graph
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_country = pd.read_excel('Country-Code.xlsx',)
df_country
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
## To check data types
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.count()
```

```python
# useful to find the count of unique value in columns
country_names = final_df.Country.value_counts().index
```

```python
country_values= final_df.Country.value_counts().values
```

```python
#Pie chart
plt.pie(country_values,labels =country_names)
```

```python
#pie chart for top 3 countries
plt.pie(country_values[:3], labels = country_names[:3],autopct='%1.2f%%')
```

# Observation:
We can observe that maximum transactions are from India then USA and then UK

```python
#more relationships
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating_Count'})
```

# Observation:
The main features of the above data are that we are getting the count of the data which have good poor excellent ratings present in the dataset

```python
ratings
```

## Observation:
1. We observe that from 4.5 to 4.9 ----> Excellent
2. We observe that from 4.0 to 4.4 ----> Very Good
3. We observe that from 3.5 to 3.9 ----> Good
4. We observe that from 2.5 to 3.4 ----> Average
5. we observe that from 1.8 to 2.4 ----> Poor
6. we observe that 0 ----> No Ratings

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y= 'Rating_Count',data=ratings)
```

```python
#Use of colors
sns.barplot(x='Aggregate rating',y= 'Rating_Count',hue='Rating color',data=ratings)
```

```python
#Mapping colors
plot=sns.barplot(x='Aggregate rating',y= 'Rating_Count',hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'],)
# plot.set_facecolor(0,0,0,0)
```
